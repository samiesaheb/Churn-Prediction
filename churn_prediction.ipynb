# Customer churn prediction, 
dataset found in the kaggle customer churn prediction 2020 competition

```python
import pandas as pd
import matplotlib.pyplot as plt
# !pip install seaborn
import seaborn as sns
```

```python
train_data = pd.read_csv('train.csv')
test_data = pd.read_csv('test.csv')
```

```python
print("Train dataset shape:", train_data.shape)
```

```python
print("First few rows of train dataset:")
print(train_data.head())
```

```python
print("Train dataset data types:")
print(train_data.dtypes)
```

```python
print("Summary statistics of train dataset:")
print(train_data.describe())
```

```python
print("Missing values in train dataset:")
print(train_data.isnull().sum())
```

```python
train_data.hist(figsize=(12, 10))
plt.tight_layout()
plt.show()
```

```python
train_data.hist(figsize=(12, 10))
plt.tight_layout()
plt.savefig('histograms.png')  # Save the figure as an image
```

```python
plt.figure(figsize=(12, 10))
sns.heatmap(train_data.corr(), annot=True, cmap='coolwarm')
plt.title('Correlation Matrix Heatmap')
plt.show()
```

```python
plt.figure(figsize=(12, 6))
train_data.boxplot(column=['total_day_minutes', 'total_eve_minutes', 'total_night_minutes', 'total_intl_minutes'])
plt.title('Box Plot of Call Minutes')
plt.show()
```

```python
plt.figure(figsize=(12, 6))
sns.countplot(x='state', hue='churn', data=train_data)
plt.title('Churn Counts by State')
plt.show()
```
